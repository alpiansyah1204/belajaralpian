#Linear regresion 1

##membuat x dan y 

```python
x = df.drop(columns=['model','transmission','fuelType','brand'])
y = df['price']
```

```python
y
```

```python
y = df.price
y
```

##membagi data training dan data testing 80:20








```python
x_train,x_test,y_train,y_test = train_test_split(x,y , test_size=0.2)
```

##model linear regresi

```python
lin_reg = LinearRegression()
```

##mencoba linear regresi menggunakan data train yang sudah dibuat 

```python
lin_reg.fit(x_train,y_train)
```

##koef = teta1(kemiringan) intercept = teta0(nilai y pada x=0)

```python
print('koefisien')
print(lin_reg.coef_)
print('intercept')
print(lin_reg.intercept_)
```

```python
coef_dict = {'feature': x.columns, 'koefisien': lin_reg.coef_}

coef = pd.DataFrame(coef_dict, columns=['feature', 'koefisien'])
coef
```

##score linear regresi

```python
lin_reg.score(x_test,y_test)
```

##membuat variable hasil dari multiple regresi linear 

```python
y_predict =  lin_reg.predict(x_test)
pd.DataFrame(y_predict)
```

```python
y_test
```

```python
from yellowbrick.regressor import PredictionError, ResidualsPlot
visualizer = PredictionError(lin_reg)

visualizer.fit(x_train, y_train)  # Fit the training data to the visualizer
visualizer.score(x_test, y_test)  # Evaluate the model on the test data
visualizer.poof()
```

```python
visualizer = ResidualsPlot(lin_reg)
visualizer.fit(x_train, y_train)  
visualizer.score(x_test, y_test) 
visualizer.poof()
```

##Cek RMSE 

```python
# Menghitung RMSE (Root Mean Squared Error)
from sklearn import metrics

print(np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```
